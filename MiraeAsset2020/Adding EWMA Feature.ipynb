```python
time_series = []
for time in stocks['기준일자'] : 
    year = str(time)[:4]
    month = str(time)[4:6]
    time_series.append(year+month)
```

```python
time_series[:5]
```

```python
stocks['기준일자'] = time_series
```

```python
stocks.head()
```

```python
stock_list = set(stocks['종목번호'])
```

```python
unit = (stocks['거래금액_만원단위']/stocks['거래량'])*10000
stocks['거래금액주가'] = unit
```

```python
stocks.head()
```

```python
open_df = pd.DataFrame(index=set(stocks['기준일자']), columns=stock_list)
open_df = open_df.sort_index()
close_df = pd.DataFrame(index=set(stocks['기준일자']), columns=stock_list)
close_df = close_df.sort_index()
high_df = pd.DataFrame(index=set(stocks['기준일자']), columns=stock_list)
high_df = high_df.sort_index()
low_df = pd.DataFrame(index=set(stocks['기준일자']), columns=stock_list)
low_df = low_df.sort_index()
vol_df = pd.DataFrame(index=set(stocks['기준일자']), columns=stock_list)
vol_df = vol_df.sort_index()
unit_df = pd.DataFrame(index=set(stocks['기준일자']), columns=stock_list)
unit_df = unit_df.sort_index()
```

```python
for stock in stock_list : 
    temp = stocks[stocks['종목번호'] == stock]

    for time in set(stocks['기준일자']) : 
        try : 
            tmp = temp[temp['기준일자'] == time]
            open_df[stock][time] = int(tmp['종목시가'].mean())
            close_df[stock][time] = int(tmp['종목종가'].mean())
            high_df[stock][time] = int(tmp['종목고가'].mean())
            low_df[stock][time] = int(tmp['종목저가'].mean())
            vol_df[stock][time] = int(tmp['거래량'].mean())
            unit_df[stock][time] = int(tmp['거래금액주가'].mean())
            
        except ValueError : 
            pass
```
